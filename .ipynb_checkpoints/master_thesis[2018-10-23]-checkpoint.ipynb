# 冷却実験解析

```python
from opmap.RawCam import RawCam
from opmap.VmemMap import VmemMap
from opmap.PhaseMap import PhaseMap
from opmap.PhaseVarianceMap import PhaseVarianceMap
from opmap.util import makeMovie
```

```python
def FFT(vmem_1d, N, dt, fc_low=None, fc_high=None):
    fq = np.linspace(0, 1.0/dt, N)
    F = np.fft.fft(vmem_1d)
    if fc_low is not None:
        F[(fq <= fc_low)] = 0
    if fc_high is not None:
        F[(fq >= fc_high)] = 0
    
    F_abs = np.abs(F)
    F_abs_amp = F_abs/N
#     F_abs_amp[0] = F_abs_amp[0]/2
    F_abs_amp[0] = 0
    
    F_ifft = np.fft.ifft(F)
    F_ifft_real = F_ifft.real * 2
    
    return F_ifft_real
```

## APD延長効果の確認

### 20181011-1/C001H001S0009

```python
raw = RawCam(path='/mnt/Jalife/ExperimentData/20181011-1/C001H001S0009/', cam_type='sa4',
             image_width=128, image_height=128, frame_start=4500, frame_end=-1)
```

```python
vmem = VmemMap(raw)
vmem.smooth(size=9)
vmem.setDiffRange(diff_min=60)
vmem.morphROI(erosion=7)

for i in range(128):
    for j in range(128):
        vmem.data[:, i, j] = FFT(vmem.data[:, i, j], 500, 0.002, 2, 8)
```

```python
raw.showFrame(0)
```

```python
v = vmem.roi
v[40, 80] = 0
v[80, 62] = 0
plt.imsave('/mnt/Omer/Project/03.LinearRegionalCooling/AnalysisResult/Experiment/20181011-1/C001H001S0009/roi.png', v)
```

```python
vmem.showFrame(frame=400)
```

```python
vmem_wo_cooling = vmem.data[100:300, 40, 80]
vmem_w_cooling = vmem.data[100:300, 80, 62]

vmem_wo_cooling = (vmem_wo_cooling-np.min(vmem_wo_cooling))/(np.max(vmem_wo_cooling) - np.min(vmem_wo_cooling))
vmem_w_cooling = (vmem_w_cooling-np.min(vmem_w_cooling))/(np.max(vmem_w_cooling) - np.min(vmem_w_cooling))
```

```python
t = np.arange(0, 200*2, 2)
plt.plot(t, vmem_w_cooling, color='b')
plt.plot(t, vmem_wo_cooling, color='g')
plt.xlim([0, 400])
plt.ylim([0, 1])

plt.title("frame:4600-4800, blue:(x,y)=(62,80), green:(x,y)=(80,40)")
plt.xlabel('Time (ms)')
plt.ylabel('Normalized membrane potential (a.u.)')


plt.savefig('/mnt/Omer/Project/03.LinearRegionalCooling/AnalysisResult/Experiment/20181011-1/C001H001S0009/APD.png', dpi=600)
plt.close()
# plt.ylim([0, 1])
```
